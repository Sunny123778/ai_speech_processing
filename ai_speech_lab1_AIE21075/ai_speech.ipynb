## **A1**

```python
import librosa
import matplotlib.pyplot as plt
y, sr = librosa.load('recordd.wav')
librosa.display.waveshow(y)
```

# **A2**

```python
import numpy as np
print("Length of Signal:",len(y),"Samples")
print("Magnitude Range:",np.min(y),"to", np.max(y))
print("Sampling Rate:",sr, "Hz")
```

# **A3**

```python
import IPython.display as ipd

start = int(1.5 * sr)
end = int(2.2 * sr)
segment = y[start:end]
librosa.display.waveshow(segment, color = 'pink')
ipd.Audio(segment, rate=sr)
```

# **A4**

```python
segment_1_start = int(1.5 * sr)
segment_1_end = int(2.0 * sr)
segment_2_start = int(2.1 * sr)
segment_2_end = int(2.5 * sr)
segment_1 = y[segment_1_start:segment_1_end]
segment_2 = y[segment_2_start:segment_2_end]
print(" Segment 1:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_1, color = 'red')
plt.title('Segment1')
plt.show()
ipd.Audio(segment_1, rate=sr)
print(" Segment 2:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_2, color = 'yellow')
plt.title('Segment2')
plt.show()
ipd.Audio(segment_2, rate=sr)
```
